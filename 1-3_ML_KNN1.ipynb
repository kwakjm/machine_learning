```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# 산점도 그리기
plt.scatter(bream_length, bream_weight, label='Bream', color='blue', marker='o', edgecolors=(0, 0, 0), s= 50)
plt.scatter(smelt_length, smelt_weight, label='Smelt', color='red', marker='o',edgecolors=(0, 0, 0), s =50)
plt.scatter(30, 600, label='Data', color='green', marker='^', edgecolors=(0, 0, 0), s =50)

# 축 레이블 설정
plt.xlabel('Length')
plt.ylabel('Weight')

# 제목 설정
plt.title('Fish Length vs. Weight', fontsize=10)

# 그리드 표시
plt.grid(True, linestyle='--', alpha=0.5)


plt.legend()



# 그래프 출력
plt.show()
```

```python
#scikit-learn이 기대하는 data set
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
print(length)
print(weight)
```

```python
fish_data = [[l,w] for l,w in zip(length ,weight)]
print(fish_data)
```

```python
# 지도학습이기 때문에 dataset의 정답(도미인지 빙어인지)을 알려줘야함.

fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()

#fit을 통해 훈련, x와 답안 y를 출력
kn.fit(fish_data, fish_target)

#성능을 보는방법(정확도, 가지고 있는 것중에 몇개를 맞추었나)
kn.score(fish_data, fish_target)
```

```python
kn.predict([[30,600]])
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn49 = KNeighborsClassifier(n_neighbors=49)

#fit을 통해 훈련, x와 답안 y를 출력
kn49.fit(fish_data, fish_target)

#성능을 보는방법(정확도, 가지고 있는 것중에 몇개를 맞추었나)
kn49.score(fish_data, fish_target)
```

```python
kn = KNeighborsClassifier()

#fit을 통해 훈련, x와 답안 y를 출력
kn.fit(fish_data, fish_target)

for n in range(5,50):
  kn.n_neighbors = n
  score = kn.score(fish_data, fish_target)

  if score <1:
    print(n ,score)
    break
```
